```python
import ast
```

```python

def convert(text):
    if isinstance(text, str):
        try:
            data = ast.literal_eval(text)
        except (ValueError, SyntaxError):
            return []  # fallback for malformed strings
    else:
        data = text  # assume it's already a Python object

    return [i['name'] for i in data if 'name' in i]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies['crew'] = movies['crew'].apply(convert)
```

```python
def fetch_directory(text):
    L = []
    # Use literal_eval only if text is a string
    if isinstance(text, str):
        try:
            crew_data = ast.literal_eval(text)
        except (ValueError, SyntaxError):
            return []
    else:
        crew_data = text

    # Extract director names
    for i in crew_data:
        if isinstance(i, dict) and i.get('job') == 'Director':
            L.append(i.get('name'))

    return L
```

```python
# Preparing the overview
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Combine all the relevant features
movies['tags'] = movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
# Creating a new data frame

new_df = movies[['movie_id','title','tags']]
# new_df
new_df.head(2)
```

```python
# COnvert the list to the string

new_df['tags'] = new_df['tags'].apply(lambda x: "".join(x))
```

```python
new_df
```

```python
# converting text into lowercase
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# Text vectorization
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

new_df.head()
```

```python
# Measure using cosine similarity

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
# Recommendation function

def recommend(movie):
   movie_index = new_df[new_df['title'] == movie].index[0]
   distance = similarity[movie_index]
   movie_list = sorted(list(enumerate(distance)), reverse=True,key=lambda x:x[1])[1:6]

   for i in movie_list:
      print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
