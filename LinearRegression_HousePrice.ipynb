```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = {'Area' : [2600,3000,3200,3600,4000],
         'Price':[550000,565000,610000,680000,725000]
      }

# Create the DataFrame
df = pd.DataFrame(data)
df
```

```python
# Step 2: Separate features (X) and target (y)
X = df[['Area']]   # Feature: Area (must be in 2D shape)
y = df['Price']    # Target: Price
```

```python
# Step 3: Linear regression object creation
model = linear_model.LinearRegression()
model.fit(X,y)
```

```python
# Make a prediction
area_to_predict = int(input("Enter area:"))
predicted_price = model.predict([[area_to_predict]])

print(f"Predicted price for {area_to_predict} sq ft = ${predicted_price[0]:,.2f}")
```

```python
# Step 5: (Optional) Plot the regression line
plt.scatter(df['Area'], df['Price'], color='blue', label='Actual Data')
plt.plot(df['Area'], model.predict(X), color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Price Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
